```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz, process
from nltk.corpus import words
import re
import string
import time
import unidecode

outputs_dir = "/Volumes/GoogleDrive/.shortcut-targets-by-id/0B0efpUDNVRiVME9hOFNHaU1xVW8/Nutrition_BiasInScience/WebOfScience/Industry_Tagging/Outputs/"
```

```python
final_long = pd.read_csv(f'{outputs_dir}wos_indtagged_final_long.csv')
```

```python
# Establish "match type priority"
# Top 100, US, UK, Suffix

def match_type(row):
    if row['Is_Industry'] != 1:
        return ""
    elif row['Is_Industry_Top100'] == 1:
        return "Top100"
    elif row['Is_Industry_US'] == 1:
        return "US"
    elif row['Is_Industry_UK'] == 1:
        return "UK"
    elif row['Is_Industry_Suf_Company'] == 1:
        return 'SuffixCompany'
    else:
        print("Error in the industry labelling")
        raise 

final_long['Match_Type'] = final_long.apply(match_type, axis=1)
```

```python
# Remove extra label columns 
final_long_trim = final_long.drop(final_long.columns[4:9].tolist() + final_long.columns[10:15].tolist(), axis = 1)
```
